#Extração dos dados

```python
import pandas as pd
import json
```

```python
with open('/content/TelecomX_Data.json') as f:
    dados = json.load(f)
```

```python
df = pd.json_normalize(dados)
```

```python
df.head()
```

```python
df.set_index('customerID', inplace=True)
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.isnull().sum()
```

#Padronização de textos


```python
colunas_categoricas = [
    'customer.gender', 'customer.Partner', 'customer.Dependents',
    'phone.PhoneService', 'phone.MultipleLines', 'internet.InternetService',
    'internet.OnlineSecurity', 'internet.OnlineBackup', 'internet.DeviceProtection',
    'internet.TechSupport', 'internet.StreamingTV', 'internet.StreamingMovies',
    'account.Contract', 'account.PaperlessBilling', 'account.PaymentMethod', 'Churn'
]
```

```python
for col in colunas_categoricas:
    df[col] = df[col].str.strip().str.lower()
```

#Converter campos numéricos

```python
df['account.Charges.Monthly'] = pd.to_numeric(df['account.Charges.Monthly'], errors='coerce')
df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors='coerce')
```

```python
print(df.isnull().sum())
```

#Preencher os gastos totais que estão em branco

```python
df.loc[df['account.Charges.Total'].isnull(),'account.Charges.Total'] = df['customer.tenure'] * df['account.Charges.Monthly']
```

```python
print(df.isnull().sum())
```

#Coluna de contas diárias

```python
df['account.Charges.Daily'] = df['account.Charges.Monthly'].apply(lambda x: x/30)
```

```python
charges_daily = df.pop('account.Charges.Daily')
```

```python
df.insert(18, 'account.Charges.Daily',charges_daily )
```

#Proporção de clientes que deixaram a empresa

```python
churn_counts = df['Churn'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
churn_counts.plot(kind='pie', autopct='%1.1f%%', colors=['green', 'red'], startangle=90)
plt.title('Proporção de evasão dos Clientes')
plt.ylabel('')
plt.show()
```

#Contagem de evasão por variáveis categóricas

```python
import seaborn as sns
plt.style.use('seaborn-v0_8')
sns.set_palette('pastel')
```

#Análise por tipo de contrato

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='account.Contract', hue='Churn')
plt.title('Evasão por Tipo de Contrato')
plt.xlabel('Contrato')
plt.ylabel('Quantidade')
plt.show()
```

#Análise por tipo de pagamento

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='account.PaymentMethod', hue='Churn')
plt.title('Evasão por Método de Pagamento')
plt.xlabel('Método de Pagamento')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.show()
```

#Análise por tempo de permanência

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='customer.tenure', bins=30, hue='Churn', multiple='stack')
plt.title('Distribuição de evasão por tempo de permanência')
plt.xlabel('Meses')
plt.ylabel('Quantidade')
plt.show()
```

#Relatório final

O presente trabalho tem como objetivo levantar dados sobre os clientes da empresa Telecom X e, a partir dessa análise, identificar possíveis causas de evasão, além de sugerir ações para solucionar o problema.

Os dados utilizados foram fornecidos pela empresa através do link “[https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json](https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json)”, armazenados em um arquivo no formato JSON. Inicialmente, realizamos conferências e correções de dados inconsistentes e criamos uma coluna para apresentar os valores diários gastos pelos clientes, permitindo uma análise mais precisa da relação entre a empresa e seus consumidores.

Na análise dos dados, buscamos padrões entre os clientes que deixaram a empresa, considerando aspectos como tempo e tipo de contrato, método de pagamento, cobrança mensal, faixa etária e adesão a serviços adicionais.

Concluímos que os clientes com maior índice de evasão:
   - Possuem contratos do tipo “month-to-month”;
   - Realizam pagamentos via cheque eletrônico;
   - São clientes recentes.

Além disso, observamos que clientes que utilizam o serviço de fibra óptica também apresentam maior propensão à evasão.

### Recomendações:

- Oferecer alternativas para que clientes com contratos “month-to-month” migrem para planos anuais, garantindo benefícios e descontos pela fidelidade.
- Identificar possíveis problemas técnicos e verificar se o processo de pagamento via cheque eletrônico é intuitivo para diferentes perfis de clientes.
- Investir em estratégias de pós-venda e pesquisas de satisfação com clientes recentes.
- Assegurar que o serviço de internet via fibra óptica atenda aos padrões de qualidade exigidos pelo mercado e que seu preço seja competitivo em relação aos concorrentes.
